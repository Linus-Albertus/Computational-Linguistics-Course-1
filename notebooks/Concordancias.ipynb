## Concordancias en corpus externo

```python
# Directorio del corpus:
corpusRoot = '/Users/lino/PycharmProjects/Proyecto1_NLTK/PRESEEA/'
# Importar todos los archivos del directorio del corpus:
corpusRaw = PlaintextCorpusReader(corpusRoot, '.*')
corpus = corpusRaw.fileids() # Lista de todos los elementos del corpus.
print('El corpus está compuesto de los siguientes archivos:')
for documento in corpus:
    print(documento)
```

### Concordancias:

```python
# Palabra a buscar:
palabra = 'espanto'
# Tamaño del contexto (en caracteres) de la concordancia:
tamanoContexto = 70
# Número de casos máximo a buscar en cada documento del corpus:
numeroCasosMaximo = 100
```

```python
for documento in corpus:
    texto = Text(corpusRaw.words(documento))
    print('---- Documento: {0} ----'.format(documento))
    texto.concordance(palabra, tamanoContexto, numeroCasosMaximo)
    print()
```

### Búsqueda de palabras similares en el corpus:

```python
corpusCompleto = Text(corpusRaw.words())
corpusCompleto.similar(palabra)
```

### Buscador por raíz (*stem*) en el corpus completo:

```python
spanish_stemmer = nltk.stem.SnowballStemmer("spanish") # Stemmer para el español.

class IndexedText(object): # Crea una clase que permite buscar concordancias utilizando un stemmer.

    def __init__(self, stemmer, text):
        self._text = text
        self._stemmer = stemmer
        self._index = nltk.Index((self._stem(word), i)
                                 for (i, word) in enumerate(text))

    def concordance(self, word, width=40):
        key = self._stem(word)
        wc = int(width/4)                # words of context
        contador = 0
        for i in self._index[key]:
            lcontext = ' '.join(self._text[i-wc:i])
            rcontext = ' '.join(self._text[i:i+wc])
            ldisplay = '{:>{width}}'.format(lcontext[-width:], width=width)
            rdisplay = '{:{width}}'.format(rcontext[:width], width=width)
            print(ldisplay, rdisplay)
            contador += 1
        print()
        print('---- El número de casos totales es:', contador, '----')

    def _stem(self, word):
        return self._stemmer.stem(word).lower()

text = IndexedText(spanish_stemmer, corpusCompleto)
text.concordance(palabra)
```

### Búsqueda por palabras relacionadas en WordNet:
En este caso se especifica una palabra en inglés, y se usa WordNet para buscar palabras relacionadas semánticamente en español. Se trata de una búsqueda un poco tosca, pero útil.

#### Búsqueda por texto individual, sin stemming.

```python
palabraEnIngles = 'work'
sinonimos = [sin.lemma_names('spa') for sin in wn.synsets(palabraEnIngles)]

for documento in corpus:
    texto = Text(corpusRaw.words(documento))
    print('---- Documento: {0} ----'.format(documento))
    print()
    for grupo in sinonimos:
        for forma in grupo:
            print('****** Palabra: {0} ******'.format(forma))
            texto.concordance(forma, tamanoContexto, numeroCasosMaximo)
            print()
```

#### Búsqueda en el corpus completo, con stemming.

```python
palabraEnIngles = 'work'
sinonimos = [sin.lemma_names('spa') for sin in wn.synsets(palabraEnIngles)]

for grupo in sinonimos:
    for forma in grupo:
        print('****** Palabra: {0} ******'.format(forma))
        text = IndexedText(spanish_stemmer, corpusCompleto) 
        text.concordance(forma)
        print()
```
