### Ansiedad y calificaciones en exámenes. Gráfico de puntos (scatterplot) con línea de regresión y error del modelo.

```r
setwd("~/PycharmProjects/doctoradoLino/R_Data/R_Book") # Directorio de trabajo.
```

```r
examData <- read.delim("Exam Anxiety.dat", header = TRUE)
```

```r
options(repr.plot.width=8, repr.plot.height=5) # Tamaño del gráfico. Se define esto para cada celda del notebook.
scatter <- ggplot(examData, aes(Anxiety, Exam, colour = Gender)) # Agrega el factor Gender a las dos variables Anxiety y Exam.
graficoExamData <- scatter + geom_point() + geom_smooth(method = "lm", aes(fill = Gender), alpha = 0.2) # Agrema modelos lineales para cada grupo (factor) y transparencias de colores al error estandar de cada modelo.
graficoExamData 
```

### Niveles de higiene en el primer día de un festival, de 0 (apesta) a 4 (huele a un jardín de rosas). Histograma con una curva normal con la media y desviación estándar de los datos.

```r
dlf <- read.delim("DownloadFestival(No Outlier).dat", header=TRUE)
```

```r
hist.day1 <- ggplot(dlf, aes(day1)) + geom_histogram(aes(y = ..density..), colour = "blue", fill = "white") + labs(x = "Higiene en el día 1", y = "Densidad") + stat_function(fun = dnorm, args = list(mean = mean(dlf$day1, na.rm = TRUE), sd = sd(dlf$day1, na.rm = TRUE)), colour = "purple", size = 1)
hist.day1
```
